```python
import pandas as pd
import pickle
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder,MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.utils import resample
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import VotingClassifier
```

```python
df=pd.read_csv("kidney_disease.csv")
```

```python
important_columns=['age','bp','sg','al','hemo','sc','htn','dm','cad','appet','pc','classification']
df=df[important_columns]
df
```

```python
df['age'].fillna(df['age'].median(),inplace=True)
```

```python
df['bp'].fillna(df['bp'].median(),inplace=True)
df['sg'].fillna(df['sg'].mode()[0],inplace=True)
df['al'].fillna(df['al'].mode()[0],inplace=True)
df['hemo'].fillna(df['hemo'].median(),inplace=True)
df['sc'].fillna(df['sc'].median(),inplace=True)
df['htn'].fillna(df['htn'].mode()[0],inplace=True)
df['dm'].fillna(df['dm'].mode()[0],inplace=True)
df['cad'].fillna(df['cad'].mode()[0],inplace=True)
df['appet'].fillna(df['appet'].mode()[0],inplace=True)
df['pc'].fillna(df['pc'].mode()[0],inplace=True)

df.isnull().sum()
```

```python
df['dm'].value_counts()
```

```python
for i in df.select_dtypes(include='object').columns:
    print(df[i].value_counts())
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip().replace('\t', '', regex=True)

print(df['cad'].value_counts())
print(df['dm'].value_counts())
print(df['classification'].value_counts())
```

```python
cols_to_encode = ['htn', 'dm', 'cad', 'appet', 'pc', 'classification']

for col in cols_to_encode:
    df[col] = df[col].str.strip().str.lower()   # clean whitespace + lowercase
    df[col] = df[col].map({
        'yes': 1, 'no': 0,
        'abnormal': 0, 'normal': 1,
        'ckd': 1, 'notckd': 0,
        'poor': 1, 'good': 0   # for appet if values are good/poor
    })

```

```python
cols_to_encode = ['htn', 'dm', 'cad', 'appet', 'pc', 'classification']

for col in cols_to_encode:
  print(df[col].value_counts())
```

```python
numeric_column=['age','bp','sg','al','hemo','sc']

scaler=MinMaxScaler()
df[numeric_column]=scaler.fit_transform(df[numeric_column])
df.head()
```

```python
df['classification'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE,RandomOverSampler
x=df.drop('classification',axis=1)
y=df['classification']

smote=SMOTE(random_state=42)
x_balanced,y_balanced=smote.fit_resample(x,y)

y_balanced.value_counts()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_balanced,y_balanced,test_size=0.2,random_state=42)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model= {
    "LogisticRegression": LogisticRegression(),
    "RandomForestClassifier": RandomForestClassifier(),
    "SVC": SVC(),
    "KNeighborsClassifier": KNeighborsClassifier(),
    "GaussianNB": GaussianNB(),
    "DecisionTreeClassifier": DecisionTreeClassifier(),
    "AdaBoostClassifier": AdaBoostClassifier(),
    "GradientBoostingClassifier": GradientBoostingClassifier()
}

for name, model in model.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy_score_value = accuracy_score(y_test, y_pred)
    classification_report_str = classification_report(y_test, y_pred)
    confusion_matrix_str = confusion_matrix(y_test, y_pred)
    print(f"Accuracy Score for {name}: {accuracy_score_value}")
    print(f"Classification Report for {name}:\n{classification_report_str}")
    print(f"confusion matrix {name}\n: {accuracy_score_value}")
```

```python
model_gbc=GradientBoostingClassifier()
model_gbc.fit(x_train,y_train)
y_pred=model_gbc.predict(x_test)
print("confusion matrix \n: ",confusion_matrix(y_test,y_pred))
print("classification report \n: ",classification_report(y_test,y_pred))
```

```python
def predict_chronic_disease(age, bp, sg, al, hemo, sc, htn, dm, cad, appet, pc):
    # 1. Build DataFrame
    df = pd.DataFrame([{
        'age': age,
        'bp': bp,
        'sg': sg,
        'al': al,
        'hemo': hemo,
        'sc': sc,
        'htn': htn,
        'dm': dm,
        'cad': cad,
        'appet': appet,
        'pc': pc
    }])

    # 2. Map categorical variables
    df['htn']   = df['htn'].map({'yes': 1, 'no': 0})
    df['dm']    = df['dm'].map({'yes': 1, 'no': 0})
    df['cad']   = df['cad'].map({'yes': 1, 'no': 0})
    df['appet'] = df['appet'].map({'good': 1, 'poor': 0})
    df['pc']    = df['pc'].map({'normal': 1, 'abnormal': 0})

    # 3. Scale **all 11 columns**, because your saved scaler expects 11 features
    df = pd.DataFrame(
        scaler.transform(df),
        columns=scaler.feature_names_in_
    )

    # 4. Predict
    prediction = model_gbc.predict(df)
    return prediction[0]


# ---- Test it ----
result = predict_chronic_disease(
    age=30, bp=20, sg=1.020, al=1.0, hemo=15.4, sc=1.2,
    htn="no", dm="no", cad="no", appet="good", pc="normal"
)

if result == 1:
    print("patient has ckd")
else:
    print("patient does not have ckd")
```
