```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
a=sns.countplot(data=df,x="Gender")
a.bar_label(a.containers[0])
plt.show()
```

```python
# """from above chart we analysed that the no. females in data is more than no. of males"""
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python

ax=gb.plot(kind='bar')
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=3)
plt.title("Average Scores by Parent Education Level")
plt.xlabel("Parent Education Level")
plt.ylabel("Average Score")
plt.show()
```

```python
sns.heatmap(gb,annot=True)
plt.show()
```

```python
# from above analysis we conclude that parents education have good impact on students scores
```

```python
gb2=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb2)
```

```python
sns.heatmap(gb2,annot=True)
plt.show()
```

```python
#from above analysis we concluded that parests marritial status has nigligible on student scores
```

```python
gA=df.loc[(df["EthnicGroup"]=="group A")].count()
gB=df.loc[(df["EthnicGroup"]=="group B")].count()
gC=df.loc[(df["EthnicGroup"]=="group C")].count()
gD=df.loc[(df["EthnicGroup"]=="group D")].count()
gE=df.loc[(df["EthnicGroup"]=="group E")].count()
l=["group A","group B","group C","group D","group E"]
mlist=[gA["EthnicGroup"],gB["EthnicGroup"],gC["EthnicGroup"],gD["EthnicGroup"],gE["EthnicGroup"]]
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethnic Group")
plt.show()
```
